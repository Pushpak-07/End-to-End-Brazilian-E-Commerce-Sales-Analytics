```python
master_df = master_df.merge(
    products,
    on="product_id",
    how="left"
)
```

```python
master_df = master_df.merge(
    sellers,
    on="seller_id",
    how="left"
)
```

```python
master_df = master_df.merge(
    payments,
    on="order_id",
    how="left"
)
```

```python
master_df = master_df.merge(
    reviews,
    on="order_id",
    how="left"
)
```

```python
master_df["Order Value"] = master_df["price"] + master_df["freight_value"]
```

```python
date_columns = [
    "order_purchase_timestamp",
    "order_delivered_customer_date",
    "order_estimated_delivery_date"
]

for col in date_columns:
    master_df[col] = pd.to_datetime(master_df[col])
```

```python
master_df["Delivery Days"] = (
    master_df["order_delivered_customer_date"]
    - master_df["order_purchase_timestamp"]
).dt.days
```

```python
master_df["Delivery Delay"] = (
    master_df["order_delivered_customer_date"]
    - master_df["order_estimated_delivery_date"]
).dt.days
```

```python
master_df["Delivery Status"] = np.where(
    master_df["Delivery Delay"] <= 0,
    "On Time",
    "Late"
)
```

```python
master_df["Purchase Month"] = master_df[
    "order_purchase_timestamp"
].dt.month_name()
```

```python
master_df["Purchase Year"] = master_df[
    "order_purchase_timestamp"
].dt.year
```

```python
master_df["Purchase Quarter"] = master_df[
    "order_purchase_timestamp"
].dt.quarter
```

```python
master_df["Customer Type"] = pd.cut(
    master_df["Order Value"],
    bins=[0,100,300,1000,100000],
    labels=[
        "Low Value",
        "Medium Value",
        "High Value",
        "Premium"
    ]
)
```

## Business Question 6
Do the top 20% of sellers generate around 80% of the total revenue?

```python
# Revenue by Seller
seller_revenue = (
    master_df.groupby("seller_id")["Order Value"]
    .sum()
    .sort_values(ascending=False)
    .reset_index()
)
```

```python
#Calculate Revenue Contribution
seller_revenue["Revenue %"] = (
    seller_revenue["Order Value"]
    / seller_revenue["Order Value"].sum()
) * 100

seller_revenue["Cumulative Revenue %"] = seller_revenue["Revenue %"].cumsum()
```

```python
# Calculate Seller Percentage
seller_revenue["Seller Rank"] = range(1, len(seller_revenue) + 1)

seller_revenue["Seller %"] = (
    seller_revenue["Seller Rank"]
    / len(seller_revenue)
) * 100
```

```python
# Find Revenue from Top 20% Sellers
top20 = seller_revenue[seller_revenue["Seller %"] <= 20]

top20_revenue = top20["Revenue %"].sum()

print(f"Top 20% sellers contribute {top20_revenue:.2f}% of total revenue.")
```

```python
plt.figure(figsize=(14,6))

# Revenue bars
plt.bar(
    seller_revenue["Seller Rank"],
    seller_revenue["Revenue %"],
    color="steelblue",
    label="Revenue %"
)

# Cumulative revenue line
plt.plot(
    seller_revenue["Seller Rank"],
    seller_revenue["Cumulative Revenue %"],
    color="red",
    linewidth=2,
    label="Cumulative Revenue %"
)

# 80% reference line
plt.axhline(
    y=80,
    color="green",
    linestyle="--",
    label="80% Revenue"
)

# 20% seller reference line
plt.axvline(
    x=len(seller_revenue) * 0.20,
    color="orange",
    linestyle="--",
    label="Top 20% Sellers"
)

plt.xlabel("Seller Rank")
plt.ylabel("Revenue Contribution (%)")
plt.title("Pareto Analysis of Seller Revenue")

plt.legend()

plt.tight_layout()

plt.show()
```

### Business Insight

The Pareto Analysis evaluates whether a small proportion of sellers contributes the majority of total revenue. If the top 20% of sellers generate approximately 80% of revenue, it confirms the Pareto Principle (80/20 Rule). This indicates that the marketplace relies heavily on a limited number of high-performing sellers, highlighting the importance of retaining and supporting these strategic partners while developing initiatives to improve the performance of lower-contributing sellers.
